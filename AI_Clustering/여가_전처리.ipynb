### csv파일을 불러와 법정동 코드와 merge

```python
def preprocessing(path):
    global df
    csv_list = os.listdir(path)
    for i in csv_list:
        col_name = i.split(' ')[1]
        temp = pd.read_csv(path + i, encoding='cp949')
        temp = temp[['영업상태명', '지번주소']]
        cond = temp['영업상태명']=='영업/정상'
        temp = temp[cond]
        temp[col_name] = temp['지번주소'].str.split(' ')[:].str.get(2)
        temp = temp[col_name].value_counts().to_frame().reset_index().rename(columns={'index':'법정동명'})
        df = pd.merge(df, temp, on='법정동명', how='outer')
    return df
```

```python
path = 'dataset/여가/'
preprocessing(path)
```
