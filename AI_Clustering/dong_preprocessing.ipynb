````python
temp = pd.read_csv('dataset/SCF_20_20220826_C_001.csv', encoding='EUC-KR')
temp
````

````python
temp2 = pd.read_excel('dataset/행정동_20220901.xlsx')
temp2
````

````python
temp.columns, temp2.columns
````

````python
temp.drop(['기준년분기코드'], axis=1, inplace=True)
temp
````

````python
temp2.drop(['생성일자', '말소일자'], axis=1, inplace=True)
temp2
````

````python
temp2['행정동코드'] = temp2['행정동코드']/100
temp2
````

````python
df = pd.merge(temp2, temp, how='inner', on='행정동코드')
df
````
