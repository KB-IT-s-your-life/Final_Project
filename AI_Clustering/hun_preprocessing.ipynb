```python
park_temp = pd.read_csv('dataset/전국도시공원정보표준데이터.csv', encoding='EUC-KR')
# 서울인 데이터만 추리기
park_temp = park_temp[park_temp['제공기관명'].str.contains('서울')]
park_temp
```

```python
park_temp['공원구분'].value_counts()
```

```python
# 어린이공원 => 동네 놀이터 // 제외
park_temp = park_temp[park_temp['공원구분'] != '어린이공원']
park_temp
```

```python
# 필요한 컬럼만 추출
park_temp = park_temp[['공원명','소재지지번주소']]
park_temp
```

```python
# 동 데이터 추출을 위해 str.split()
park_list = park_temp['소재지지번주소'].str.split(' ')
park_list
```

```python
# '시군구명', '법정동명' 컬럼 추가
park_temp['시군구명'] = park_list.str.get(1)
park_temp['법정동명'] = park_list.str.get(2)
park_temp
```

```python
# 데이터프레임으로 만들기
park = park_temp['법정동명'].value_counts().to_frame()
park.reset_index(inplace=True)
park.columns = ['법정동명', '공원수']
park
```

```python
# 서울시 노래연습장업 인허가 정보
import chardet
import pandas as pd
filename = 'dataset/서울시 노래연습장업 인허가 정보.csv'
with open(filename, 'rb') as f:
    result = chardet.detect(f.readline())  # or read() if the file is small.
    print(result['encoding'])
```

```python
# 동 데이터 추출을 위해 str.split()
sr_list = sr_temp['지번주소'].str.split(' ')
sr_list
```

```python
# '시군구명', '법정동명' 컬럼 추가
sr_temp['시군구명'] = sr_list.str.get(1)
sr_temp['법정동명'] = sr_list.str.get(2)
sr_temp
```

```python
# 데이터프레임으로 만들기
sr = sr_temp['법정동명'].value_counts().to_frame()
sr.reset_index(inplace=True)
sr.columns = ['법정동명', '노래방수']
sr
```
