```python
df.columns
```

```python
df_x = df.drop(['ADSTRD_CD', 'VSTFC_FCLTY_CNT', 'PBOFC_CNT', 'BANK_CNT', 'GNHSP_CNT',
       'HSPTL_CNT', 'PHRM_CNT', 'KNDGT_CNT', 'ELESCH_CNT', 'MSKUL_CNT',
       'HGSCHL_CNT', 'UNVRS_CNT', 'DMST_CNT', 'SUPMK_CNT', 'THTER_CNT',
       'ARPRT_CNT', 'RLROAD_STATN_CNT', 'BUS_TMNL_CNT', 'SUBWAY_STATN_CNT',
       'BUS_STATN_CNT'], axis=1)
df_o = df.drop(['ADSTRD_CD', 'VSTFC_FCLTY_CNT', '관공서_은행', '의료', '쇼핑', '여가', '교통'], axis=1)
```

```python
df_x.corr()
```

```python
sns.heatmap(df_x.corr(), annot=True)
```

```python
from sklearn.preprocessing import StandardScaler
data_scaled = StandardScaler().fit_transform(df_x)
data_scaled1 = StandardScaler().fit_transform(df_o)
```

```python
pd.DataFrame(data_scaled).describe().round(2)
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
pc = pca.fit_transform(data_scaled)
```

```python
pd.DataFrame(pc)
```

```python
plt.figure()
plt.scatter(pd.DataFrame(pc)[0], pd.DataFrame(pc)[1], alpha= 0.5)
```

### K-means

```python
from sklearn.cluster import KMeans
km = KMeans(n_clusters=4, random_state=5)
```

```python
km.fit_transform(pc)
km.labels_
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(km.labels_)
plt.title('K-Means Clustering', fontsize=15)
plt.show()
```

```python
from sklearn.metrics import silhouette_samples, silhouette_score
score = silhouette_score(pc, km.labels_)
score
```

```python
pcd = pd.DataFrame(pc)
pcd['cluster_id'] = km.labels_
pcd
```

```python
plt.figure()
sns.scatterplot(pcd[0], pcd[1], hue=pcd['cluster_id'] , alpha= 0.5)
```
