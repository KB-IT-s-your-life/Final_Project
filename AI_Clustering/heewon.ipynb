## 행정동 코드명, 시군구 코드명 추가

```python
temp = pd.read_csv('dataset/서울시 우리마을가게 상권분석서비스(구_상권영역).csv')
temp
```

```python
temp2 = pd.read_csv('dataset/서울시 우리마을가게 상권분석서비스(상권영역).csv', encoding='cp949')
temp2
```

```python
temp.columns, temp2.columns
```

```python
temp.columns = ['상권_구분_코드', '상권_구분_코드_명', '상권_코드', '상권_코드_명', '엑스좌표_값',
        '와이좌표_값', '시군구_코드', '행정동_코드', '기준_년월_코드', '형태정보']
temp
```

```python
area = pd.concat([temp, temp2], ignore_index=True)
area
```

```python
area = area[['상권_코드', '상권_코드_명', '시군구_코드', '행정동_코드']]
area
```

```python
area2 = pd.read_csv('dataset/서울시 우리마을가게 상권분석서비스(행정동별 상권변화지표).csv', encoding='cp949')
area2
```

```python
area2['행정동_코드'].value_counts()
```

```python
area2 = area2[area2['기준_년_코드']==2022][['행정동_코드', '행정동_코드_명']]
area2
```

```python
area = pd.merge(area, area2, on='행정동_코드')
area
```

```python
# 서울특별시 시군구 코드, 시군구 명
dic = {11680: '강남구', 11740: '강동구', 11305: '강북구', 11500: '강서구', 11620: '관악구', 11215: '광진구', 11530: '구로구',
        11545: '금천구', 11350: '노원구', 11320: '도봉구', 11230: '동대문구', 11590: '동작구', 11440: '마포구', 11410: '서대문구',
        11650: '서초구', 11200: '성동구', 11290: '성북구', 11710: '송파구', 11470: '양천구', 11560: '영등포구', 11170: '용산구',
        11380: '은평구', 11110: '종로구', 11140: '중구', 11260: '중랑구'}
```

```python
area['시군구_코드_명'] = area['시군구_코드'].apply(lambda x: dic.get(x, np.nan))
area
```

```python
temp = pd.read_csv('dataset/서울시_우리마을가게_상권분석서비스(구_상권_집객시설).csv', encoding='cp949')
temp2 = pd.read_csv('dataset/서울시_우리마을가게_상권분석서비스(구_상권배후지_집객시설).csv')
```

```python
cond = (temp['기준_년_코드'] == 2021) & (temp['기준_분기_코드'] == 3)
cond2 = (temp2['기준_년_코드'] == 2021) & (temp2['기준_분기_코드'] == 4)
temp = temp.loc[cond]
temp2 = temp2.loc[cond2]
display(temp, temp2)
```

```python
df = pd.concat([temp, temp2], ignore_index=True)
df
```

```python
df.columns
```

```python
df = pd.merge(df, area, how='left')
df
```

```python
df.drop(['기준_년_코드', '상권_구분_코드', '상권_구분_코드_명', '시군구_코드', '행정동_코드', '기준_분기_코드'], axis=1, inplace=True)
df
```

```python
df.isna().sum()
```

```python
df['상권_코드'].value_counts()
```

```python
df = df.drop_duplicates(['상권_코드'], keep='last')
```

```python
df['상권_코드'].value_counts()
```

```python
df = df.groupby(['시군구_코드_명', '행정동_코드_명', '상권_코드', '상권_코드_명']).mean()
df
```

```python
df.fillna(0, inplace=True)
df
```
