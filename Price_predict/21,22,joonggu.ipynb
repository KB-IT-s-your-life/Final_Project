```python
import pandas as pd
```

```python
df = pd.read_csv("data/서울시 부동산 전월세가 정보_2021.csv",encoding='cp949')
df1 = pd.read_csv("data/서울시 부동산 전월세가 정보_2022.csv",encoding='cp949')
df_sub = df.loc[df['전월세 구분']=='월세']
df1_sub = df1.loc[df['전월세 구분']=='월세']
```

```python
df_sub = pd.concat([df_sub,df1_sub])
```

```python
songpa = df_sub.loc[df_sub['자치구명']=='중구']
songpa
```

```python
songpa['신규갱신여부'].value_counts()
```

```python
songpa_gangsin = songpa.loc[songpa['신규갱신여부']=='갱신']
```

```python
songpa_gangsin = songpa_gangsin.drop(['지번구분', '지번구분명', '본번', '부번', '건물명', '계약일', '전월세 구분', '계약기간', '신규갱신여부', '계약갱신권사용여부', '자치구코드', '자치구명', '법정동코드'], axis=1)
```

```python
songpa_gangsin = songpa_gangsin.loc[songpa_gangsin['건물용도']!='아파트']
```

```python
songpa_dum = pd.get_dummies(data = songpa_gangsin, columns=['법정동명'])
songpa_dum = pd.get_dummies(data = songpa_dum, columns=['건물용도'])
songpa_dum
```

```python
songpa_dum['층'] = songpa_dum['층'].fillna(1)
songpa_dum
```

```python
songpa_dum = songpa_dum.dropna()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(songpa_dum.drop('임대료(만원)', 1), songpa_dum['임대료(만원)'], random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
model = LinearRegression().fit(x_train, y_train)
pred = model.predict(x_test)
mean_squared_error(y_test, pred)
```
