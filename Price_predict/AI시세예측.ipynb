```python
#필요없는 컬럼 drop
df=df.drop(['자치구코드','법정동코드','지번구분','지번구분명','본번','부번','계약기간', '신규갱신여부', '계약갱신권사용여부'],1)
df.head()
```

```python
#건물용도에 아파트인 것들을 제외
df = df[df['건물용도']!='아파트']
```

```python
df1 = df[df['전월세 구분']== '전세']
df2 = df[df['전월세 구분']== '월세']
```

```python
#종전보증금, 종전 임대료 nan값 제외, 0값제외
df = df[df['종전 보증금'].notna()]
df = df[df['종전 임대료'].notna()]
df = df[(df['종전 보증금']!=0) & (df['종전 임대료']!=0)]
df
```
