# 임대차 3법
- (1) '계약갱신 청구권제', (2) '전월세 상한제', (3) '전월세 신고제' 등 3개 제도를 핵심으로 하는 '주택임대차 보호법'과 '부동산 거래신고 등에 관한 법률'에 대한 개정안.
- 2020년 7월 30일에 국회 통과, (1) '계약갱신 청구권제', (2) '전월세 상한제'는 **2020년 7월 31일**에 바로 시행, (3) '전월세 신고제'는 **2021년 6월 1일**에 시행 

###  1. 계약갱신 청구권제
- 임차인이 원할 시 1회(2년 보장)에 한하여 계약 연장이 가능하다. 기존2년에서 4년으로 2년 연장을 보장.

###  2. 전월세 상한제
- **재계약을 할 때 임대료를 이전 계약의 5% 이상 올리지 못하게 하되**, 지자체마다 조례를 통해서, 이 한도 내에서 다시 상한을 정할 수 있는 제도

###  3. 전월세 신고제
- 임대차 계약 시에 30일 이내에 계약당사자(임대인,세입자)가 보증금,임대 기간등 계약관련 정보를 관청에 신고해야함
